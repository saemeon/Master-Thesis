# Chin Multivariat

```python
C = Chin_MV(intercept = True, direct = True)
pprint.pprint(C.run(10), width=1)
```

```python
"""results0 = {}
results0["t0"] ={}
results0["tW"] ={}
results0["tE"] ={}
results0["t0"]["AnaEst1tot"] =[]
results0["tW"]["AnaEst1tot"] =[]
results0["tE"]["AnaEst1tot"] =[]
results0["t0"]["NaivEst"] =[]
results0["tW"]["NaivEst"] =[]
results0["tE"]["NaivEst"] =[]
"""
for i in tqdm(range(50)):
    C = Chin_MV(intercept = True, direct = True)
    result = C.run(100)
    results0["t0"]["AnaEst1tot"].append(result["AnaEst1tot"][0])
    results0["t0"]["NaivEst"   ].append(result["NaivEst"   ][0])
    results0["tW"]["AnaEst1tot"].append(result["AnaEst1tot"][1])
    results0["tW"]["NaivEst"   ].append(result["NaivEst"   ][1])
    results0["tE"]["AnaEst1tot"].append(result["AnaEst1tot"][2])
    results0["tE"]["NaivEst"   ].append(result["NaivEst"   ][2])
```

```python
ana_mean = np.round((np.mean(results0["t0"]["AnaEst1tot"]),np.mean(results0["tW"]["AnaEst1tot"]),np.mean(results0["tE"]["AnaEst1tot"])),2)
ana_std  = np.round((np.std( results0["t0"]["AnaEst1tot"]),np.std( results0["tW"]["AnaEst1tot"]),np.std( results0["tE"]["AnaEst1tot"])),2)
naiv_mean = np.round((np.mean(results0["t0"]["NaivEst"]),np.mean(results0["tW"]["NaivEst"]),np.mean(results0["tE"]["NaivEst"])),2)
naiv_std  = np.round((np.std( results0["t0"]["NaivEst"]),np.std( results0["tW"]["NaivEst"]),np.std( results0["tE"]["NaivEst"])),2)

print("Ana mean:\t",ana_mean,"\tAna std:\t",ana_std)
print("Naiv mean:\t",naiv_mean,"\tNaiv std:\t",naiv_std)
```
